```python
# Install all libs as needed
!pip install praw pandas openpyxl
import praw
import pandas as pd
```

```python
# API Info
reddit = praw.Reddit(
    client_id='Your client_id',
    client_secret='Your client_secret',
    user_agent='Your user_agent via Reddit',
    username='Your Reddit Username',
    password='Your Reddit password')
```

```python
# Your interested Subreddit list
subreddit_list = ["Subreddit 1","Subreddit 2","..."]
```

```python
# Actual Subreddit Rule Collection Starts Here
data = []
for sub_reddit in subreddit_list:
    subreddit = reddit.subreddit(sub_reddit)
    print(sub_reddit)
    for rule in subreddit.rules:
        rule_data = {
            "Subreddit": subreddit.display_name,
            "Subreddit Description": subreddit.public_description,
            "Rule Header": rule.short_name,
            "Rule Description": rule.description,
            "Time Created": rule.created_utc,
            "Rule Violation Reason": rule.violation_reason,
        }
        data.append(rule_data)

df = pd.DataFrame(data)
df
```
